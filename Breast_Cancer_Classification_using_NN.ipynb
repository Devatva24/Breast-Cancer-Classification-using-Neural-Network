```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Separating the features and target

```python
X = data_frame.drop(columns = "label", axis = 1)
y = data_frame["label"]
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Standardize the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
X_train_std
```

**Building the neural network**

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of neural network
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)), #to convert the data (numerical into a single dimensional array) [number of input nodes = number of features or cols in the dataset]
    keras.layers.Dense(20, activation = "relu"), #this is the hidden layer where dense denotes that all the each of the nodes are connected to previous nodes
    keras.layers.Dense(2, activation = "sigmoid") #this is the output layer which is also dense [number of output node = number of classes we have ]
])
```

For label encoding like numerical one we use sparse_categorical_crossentropy  
whereas for one hot encoding we use categorical_crossentropy  
one hot encoding :  
dog = [1, 0 , 0]  
cat = [0, 1, 0]  
horse = [0, 0, 1]

```python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(X_train, y_train, validation_split = 0.1, epochs = 10)
```

training data accuracy, training data loss, validation data accuracy, validation data loss  
loss function : difference between true value and predicted value, it should be minimum

```python
history = model.fit(X_train_std, y_train, validation_split = 0.1, epochs = 10)
```

Visualizing the accuracy and loss

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("model accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(["train", "validation"], loc = "upper left")
plt.show()
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(["train", "validation"], loc = "upper right")
plt.show()
```

Accuracy of the model on the test data

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
```
